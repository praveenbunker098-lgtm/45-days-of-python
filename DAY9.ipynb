```python
import numpy as np
l = [1,2,3]
print(l)


#arr id
def array(arr):
  print(np.array(arr))

array(l)
arr = np.array(l)
print(arr)


#2D array example

arr = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])

# print(arr)
# print(arr[0][0])   # 10
# print(arr[1][2])   # 60
# print(arr[2][1])   # 80
# print(arr[1,1])
# print(arr.shape)  #Shape Check Karna
# print(arr[0:3,0])  #specific column ko acces krne k liye
# print(arr[:,0])  #column ko access krne k liye
arr[1,1] = 100
print(arr)


# 2d list example
l = [
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
]
# print(l[0])
# print(l[0][0])
l[1][1]= 100
print(l)
```

```python
# COMPARSION
# LIST
import time
start = time.time()
l = [i*2 for i in range(10000000)]
end = time.time()
print("list output:",end-start)

# ARRAY
import numpy as np
start = time.time()
arr = np.arange(10000000)
arr = arr*2
end = time.time()
print("array output:",end-start)
```

# Vector, Matrix,***bold text*** Tensor

1.   vector(1d)
2.   matrix(2d)
3.  tensor(3d)



```python
import numpy as np


# vector(1d) list
vector = [10, 20, 30]
print(vector)

# vector(1d) array
import numpy as np
arr = np.array(vector)
print(arr)

#matrix(2d) list
matrix = [
    [1, 2, 3],
    [4, 5, 6]
]
print(matrix)

#matrix(2d) array





# tensor 3d list
tensor = [
    [
        [1, 2],
        [3, 4]
    ],
    [
        [5, 6],
        [7, 8]
    ]
]
print(tensor)
```

```python
# array

import numpy as np
arr = np.arange(12).reshape(3,4)
print("shape:", arr.shape)
print("shape:", np.shape(arr))
print("dimension:", arr.ndim)
print("size:", arr.size)
print("dtype:", arr.dtype)
```
