# Class aur Object
* Line 1: Car naam ki class banayi jo ek blueprint hai, aur isme __init__ constructor hai jo har car ka naam aur color set karega.

* Line 2: Humne car1 aur car2 naam ke do real objects banaye jo is blueprint ko use kar rahe hain.

* Line 3: Python mein self ka matlab hota hai "woh object jo abhi baat kar raha hai".

```python
# Class (Blueprint)
class Car:
    def __init__(self, name, color):
        self.name = name        # Property
        self.color = color      # Property

    def start_engine(self):     # Behavior (Method)
        return f"{self.name} engine has started."

# Object (Real Entity)
car1 = Car("Swift", "Red")
car2 = Car("Ferrari", "Black")

print(car1.start_engine())  # Output: Swift ki engine start ho gayi!
print(car2.start_engine())  # Output: Ferrari ki engine start ho gayi!
```

# 2. Encapsulation (Data Hiding)

* Line 1: Python mein kisi variable ke aage double underscore __ lagane se woh private ho jata hai.

* Line 2: __price ko koi bhi class ke bahar se direct change nahi kar sakta (car.__price likhoge toh error aayega).

* Line 3: Agar price badalna hai, toh class ke andar ke method (set_price) ke through hi badal sakte ho check karke.

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance  # Private variable (Encapsulated)

    # Getter method - balance dekhne ke liye
    def get_balance(self):
        return self.__balance

    # Setter method - balance change karne ke liye rule ke sath
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

account = BankAccount("Rahul", 5000)
# account.__balance = 100000  # Yeh kaam NAHI karega (Error ya ignore hoga)
account.deposit(2000)         # Yeh sahi tarika hai
print(account.get_balance())  # Output: 7000
print(account.owner)          # Output: Rahul
print(account.deposit)
```
